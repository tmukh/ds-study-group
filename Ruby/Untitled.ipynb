```python
import pandas as pd
import numpy as np
import seaborn
import scipy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sky.csv')
```

```python
df
```

```python
df = df[['ra','dec','redshift','class']]
```

```python
f, ax = plt.subplots(figsize=(7, 7))
ax.set(xscale="log")
res = seaborn.boxplot(x=df['redshift'],y=df['class'], ax=ax)
plt.show()
```

```python
res = seaborn.catplot(x=df['redshift'],y=df['class'],alpha=0.1)
plt.xscale("log")
plt.show()
```

```python
res = seaborn.violinplot(x=df['redshift'],y=df['class'])
plt.xscale("log")
plt.show()
```

```python
df_copy = df[['class','redshift']]
df_copy['class']= df_copy['class'].map({'STAR':0,'QSO':1,'GALAXY':2})
```

```python
df_for_stars = df[['ra','dec','redshift','class']]
df_for_stars
seaborn.lmplot(x='ra', y='dec', data=df_for_stars, hue='class', fit_reg=True, palette='dark', aspect=2)
```
